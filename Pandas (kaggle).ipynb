## loc

#### The second paradigm for attribute selection is the one followed by the loc operator: label-based selection. In this paradigm, it's the data index value, not its position, which matters.

```python
wine_reviews.loc[:, ['Address', 'Rooms', 'Type']]
```

```python
wine_reviews.Suburb == 'Abbotsford'
```

```python
wine_reviews.loc[wine_reviews['Suburb'] == 'Abbotsford']
```

```python
wine_reviews[wine_reviews['Suburb']=='Abbotsford']
```

```python
wine_reviews

```

```python
wine_reviews.loc[(wine_reviews.Suburb == 'Abbotsford') & (wine_reviews.Rooms>=3)]
```

```python
wine_reviews.loc[(wine_reviews.Suburb == 'Abbotsford') | (wine_reviews.Rooms>=3)]
```

```python
wine_reviews.loc[wine_reviews.Suburb.isin(['Abbotsford', 'Williamstown'])]

# isin is lets you select data whose value "is in" a list of values.
```

```python
wine_reviews.loc[wine_reviews.Car.notnull()]
#These methods let you highlight values which are (or are not) empty (NaN)
```

#### A map is a term, borrowed from mathematics, for a function that takes one set of values and "maps" them to another set of values. In data science we often have a need for creating new representations from existing data, or for transforming data from the format it is in now to the format that we want it to be in later. Maps are what handle this work, making them extremely important for getting your work done!

```python
wine_review_Rooms_mean = wine_reviews.Rooms.mean()
wine_reviews.Rooms.map(lambda p: p - wine_review_Rooms_mean)
```

```python
reviews = pd.read_csv("winemag-data-130k-v2.csv",index_col=0)
```

```python
def stars(row):
    if row.country == 'Canada':
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1

star_ratings = reviews.apply(stars, axis='columns')
star_ratings
```

## Groupwise analysis.............................................................

One function we've been using heavily thus far is the value_counts() function. We can replicate what value_counts() does by doing the following:

```python
reviews.groupby('points').points.count()
```

```python
reviews.groupby('points').price.min()
```

```python
reviews.groupby('country').apply(lambda df: df.description.iloc[0])
```

```python
gk = reviews.groupby('country')
gk.first()
```

```python
# Finding the values contained in the "India" group
gk.get_group('India')
```

```python
# First grouping based on "Country"
# Within each team we are grouping based on "Province"
gkk = reviews.groupby(['country', 'province'])
  
# Print the first value in each group
gkk.first()
```

```python
reviews.groupby(['country']).price.agg([len, min, max])

#Another groupby() method worth mentioning is agg(),
#which lets you run a bunch of different functions on your DataFrame simultaneously.
```

```python
# A multi-index differs from a regular index in that it has multiple levels
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
countries_reviewed.reset_index()
```

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
countries_reviewed.sort_index()
```

```python
countries_reviewed.sort_values(by=['country', 'len'])
```
